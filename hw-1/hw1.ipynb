```python
#importing all the required packages like numpy,pandas,seaborn matplotlib.pyplot etc 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#here we are loading the data set using the url.using df as dataframes to read the file
header = ["variance", "skew", "curtsies", "entropy", "classes"]
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt"
df = pd.read_csv(url, names=header)
```

```python
# importing train_test_split from sklearn.model_selection which is a free machine learning library
from sklearn.model_selection import train_test_split 
```

```python
#assigning parameters x,y to data set
X = df.drop(['classes'], axis=1)
y = df['classes']
```

```python
#test_size() is used to test the required percentage of data that is between (0.1,1.0) cant be more than for 1 for the combined value for test size and train. random_state() is used to use the same random data all the time.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=25)
```

```python
#Perceptron is used to find the accurate data or not.
from sklearn.linear_model import Perceptron

ppn = Perceptron(max_iter=40, eta0=0.1, random_state=25)
ppn.fit(X_train, y_train)
```

```python
# predict is used to lable the data set.
y_pred = ppn.predict(X_test)
```

```python
y_test
```

```python
# importing metrics to to find the accuracy of the data by using accuracy_score() method
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```
